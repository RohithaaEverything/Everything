```python
#2.number is even or odd in given queue
queue = []
elements = list(map(int,input().split()))
for element in elements:
    queue.append(element)
for x in range(len(queue)):
    n = queue.pop(0)
    if ( n % 2 == 0):
        print("EVEN")
    else:
        print("ODD")
```
