```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def insert(root,data):
    if root==None:
        return Node(data)
    if(root.data<data):
        root.right=insert(root.right,data)
    else:
        root.left=insert(root.left,data)
    return root
def levelorder(root):
    queue=[]
    queue.append(root);
    while(len(queue)!=0):
        top=queue.pop(0)
        print(top.data)
        if(top.left!=None):
            queue.append(top.left)
        if(top.right!=None):
            queue.append(top.right)
def inorder(root):
    if(root==None):
        return
    inorder(root.left)
    print(root.data)
    inorder(root.right)
def preorder(root):
    if root==None:
        return
    print(root.data)
    preorder(root.left)
    preorder(root.right)
def postorder(root):
    if(root==None):
        return
    postorder(root.left)
    postorder(root.right)
    print(root.data)
def search(root,key):
    if(root==None):
        return False
    if(root.data==key):
        return True
    if(root.data<key):
        return search(root.right,key)
    return search(root.left,key)
def minVal(root):
    temp=root
    minVal=temp.data
    while(temp.left!=None):
        minVal=temp.data
        temp=temp.left
    return minVal
def delete(root,key):
    if(root==None):
        return root
    if(root.data>key):
        root.left=delete(root.left,key)
    elif(root.data<key):
        root.right=delete(root.right,key)
    else:
        if(root.left==None):
            return root.right
        if(root.right==None):
            return root.left
        root.data=minVal(root.right)
        root.right=delete(root.right,root.data)
    return root
root=None
root=insert(root,5)
root=insert(root,3)
root=insert(root,2)
root=insert(root,7)
root=insert(root,8)
print("Level Order:")
print(levelorder(root))
print("In Order:")
print(inorder(root))
print("Pre Order:")
print(preorder(root))
print("Post Order:")
print(postorder(root))
if(search(root,13)):
    print("Key is present")
else:
    print("Key is not present")
print("Before deletion:")
print("Inorder:")
inorder(root)
delete(root,5)
print("After deletion:")
print("Inorder:")
inorder(root)
```
